```python
import requests
from bs4 import BeautifulSoup

# URL of the contribution list to scrape
url = "https://indico.jlab.org/event/459/contributions/"

# Send a GET request to the URL and get the page content
response = requests.get(url)
content = response.content

# Use BeautifulSoup to parse the page content
soup = BeautifulSoup(content, "html.parser")


contrib_tags = soup.find_all('a', {"class": "js-mathjax"})
```

```python
def get_attachment(url, url_prefix):
   
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    attach = soup.find_all('a',  attrs={'class':['attachment icon-file-presentation', 'attachment icon-file-pdf']})

    attach
    if not attach:
        return ''
    return url_prefix + attach[0]['href']

  
```

```python
contrib_data = []
href_prefix='https://indico.jlab.org'


for contrib_tag in contrib_tags:
    link = href_prefix + contrib_tag['href']
    contrib_id = link.split('/')[-2]
    title = contrib_tag.text.strip().split('\n')[1].strip()
    attach_link = get_attachment(link, href_prefix)
#     print(contrib_id, link, title, attach_link)
    
    # append the contribution data to the list
    contrib_data.append((contrib_id, link, title, attach_link))

# print the contribution data in tabular form
print('{:<10} {:<60} {} {}'.format('ID', 'Page Link', 'Title', 'Attachment Link'))
for contrib in contrib_data:
    print('{:<10} {:<60} {} {}'.format(*contrib))
```

```python
# define the output file path
import csv
output_path = 'chep23_contrib_data.csv'

# write the contribution data to the output file
with open(output_path, 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['ID', 'Link', 'Title'])
    for contrib in contrib_data:
        writer.writerow(contrib)
```
